```python
# vinhos abaixo de 100
wines.loc[wines['price'] < 100]['quantity'].sum()
```

```python
# vinhos acima de 100
wines.loc[wines['price'] > 100]['quantity'].sum()
```

```python
wines['date'] = pd.to_datetime(wines['date'])
```

```python
# dt é atributo do pandas para se referir a data | wines['date']dt.month==12 retorna o mês 12
wines['date'] = pd.to_datetime(wines['date'])
wines.loc[(wines['style'] == 'Sparkling') & (wines['date'].dt.month==12)]['quantity'].sum()
```

```python
wines.loc[((wines['style'] == 'White') | (wines['style'] == 'Rose'))
         & (wines['date'].dt.month.isin([1,2,3]))]['quantity'].sum()
```

```python
wines['sub-total'] = wines['price'] * wines['quantity']

fat = f'R${wines.loc[wines["date"].dt.month < 4]["sub-total"].sum():_.2f}'
print(fat)
```

```python
wines['date'] = wines['date'].astype(str)
wines.loc[((wines['date'] >= '2021-06-21') & (wines['date'] <= '2021-09-20'))
         & (wines['style'] == 'Red')]['quantity'].sum()
```

```python
wines['date'] = pd.to_datetime(wines['date'])

fat_mes = wines.groupby(wines['date'].dt.month)['sub-total'].sum().sort_values(ascending=False)
print(fat_mes)
```

```python
ticket_medio = wines.groupby(wines['date'].dt.month)['sub-total'].sum()
wines.groupby(wines['date'].dt.month)['quantity'].sum()

ticket_medio = ticket_medio.sort_values(ascending=False)
ticket_medio.index.name = 'Mês'
ticket_medio
```
