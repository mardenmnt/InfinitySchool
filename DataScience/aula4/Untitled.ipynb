### Formatando tabela

```python
df2 = pd.DataFrame(
    {
        'Linguagem': ['Python', 'R', 'SQL', 'C', 'Java', 'JavaScript', 'Elixir'],
        'Popularidade': ['Alta', 'Baixa', 'Altíssima', 'Média', 'Altíssima', 'Altíssima', 'Baixa'],
        'Versão': ['3,10', '4,2', '5,5', '2,2', '11', '16', '1,12'],
        'Desenvolvedores': [7_000_000, 500_000, 20_000_000, 3_500_000, 16_000_000, 25_000_000, 200_000],
        'Data de Lançamento': ['1991-02-20', '1993-08-03', '1986-01-01',
                               '1972-01-01', '1995-05-23', '1995-12-04', '2012-01-01']
    }
)
df2
```

```python
df2.dtypes
```

```python
df2['Data de Lançamento'] = pd.to_datetime(df2['Data de Lançamento'])
df2.dtypes
```

```python
df2['Data de Lançamento'] = df2['Data de Lançamento'].map(lambda dt: dt.strftime('%d-%m-%Y'))
df2
```

```python
df2['Versão'] = df2['Versão'].map(lambda txt: txt.replace(',', '.'))
df2
```

```python
df2['Versão'] = df2['Versão'].map(lambda txt: float(txt))
df2.dtypes

# OU #

# df2['Versão'] = df2['Versão'].astype(float)
```
